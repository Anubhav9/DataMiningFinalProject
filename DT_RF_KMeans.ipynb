```python
X=dataset.iloc[:,0:25]
```

```python
y=dataset.iloc[:,25]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.8, random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier=classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
from  sklearn.metrics import confusion_matrix
```

```python
cf=confusion_matrix(y_test,y_pred)
```

```python
cf
```

```python
from sklearn.tree import DecisionTreeClassifier

classification=DecisionTreeClassifier()

classification=classification.fit(X_train,y_train)
```

```python
y_pred1=classification.predict(X_test)
```

```python
y_pred1
```

```python
cf1=confusion_matrix(y_test,y_pred1)
```

```python
cf
```

```python
#decision tree model

from sklearn.ensemble import RandomForestClassifier

classifier2=RandomForestClassifier(n_estimators=200, random_state=0)

classifier2=classifier2.fit(X_train,y_train)
```

```python
y_pred2=classifier2.predict(X_test)
```

```python
y_pred2
```

```python
cf2=confusion_matrix(y_test,y_pred2)
```
